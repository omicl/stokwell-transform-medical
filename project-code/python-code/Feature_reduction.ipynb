```python
import numpy as np
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import StandardScaler

ID = pd.read_csv('/content/drive/MyDrive/_end.csv')
features_matrix = np.load('/content/drive/MyDrive/features_matrix.npy')

# Alternatively, you can use the T attribute of the numpy array
features_matrix_transposed = features_matrix.transpose()
X = features_matrix_transposed

# Get label vector y (assuming it's stored in ID)
#df = ID.drop(['id','status'], axis=1)
df = ID.drop(['id','status'], axis=1)
y = df.values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X,y)

# انجام تجزیه مقادیر تکین (SVD)
U, S, Vt = np.linalg.svd(X_scaled, full_matrices=False)

# انتخاب تعداد مورد نظر از مؤلفه‌ها (برای مثال، 128 مؤلفه)
n_components = 128
X_svd = np.dot(U[:, :n_components], np.diag(S[:n_components]))

# Print the selected features
print(X_svd)
```

```python
np.save('/content/drive/MyDrive/SSVD_128.npy',X_svd)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

# Load ID and feature matrix
ID = pd.read_csv('/content/drive/MyDrive/_end.csv')
features_matrix = np.load('/content/drive/MyDrive/features_matrix.npy')

# Alternatively, you can use the T attribute of the numpy array
features_matrix_transposed = features_matrix.transpose()
X = features_matrix_transposed

# Get label vector y (assuming it's stored in ID)
#df = ID.drop(['id','status'], axis=1)
df = ID.drop(['id','status'], axis=1)
y = df.values


# Create a LinearRegression object
non_zero_cols = np.any(X != 0, axis=0)
X_non_zero = X[:, non_zero_cols]

# Create a LinearRegression object
linear_reg = LinearRegression()

# Create the RFE (Recursive Feature Elimination) object
n_features_to_select = 128  # تعداد ویژگی‌های مورد نظر خود را تعیین کنید
rfe = RFE(estimator=linear_reg, n_features_to_select=n_features_to_select)

# Fit the RFE to the data and transform the data to include only the selected features
X_rfe = rfe.fit_transform(X_non_zero, y)

# Print the selected features
print(X_rfe)

```

```python
np.save('/content/drive/MyDrive/RFE_TEST.npy', X_rfe)
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA

ID = pd.read_csv('/content/drive/MyDrive/_end2.csv')
features_matrix = np.load('/content/drive/MyDrive/features_matrix.npy')


features_matrix_transposed = features_matrix.transpose()
X=features_matrix_transposed
y = ID.values

n_components = 128

# ایجاد یک شیء PCA با تعداد مؤلفه‌های دلخواه
pca = PCA(n_components=n_components)

# اعمال PCA بر روی داده‌های خود و تبدیل آنها به تعداد مؤلفه‌های انتخاب شده
X_pca = pca.fit_transform(X,y)

# Print the selected features
print(X_pca)
```

```python
np.save('/content/drive/MyDrive/PCA2.npy', X_pca)
```
